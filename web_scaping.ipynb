```
from urllib.request import urlopen
from bs4 import BeautifulSoup as BS
import datetime

result = []

url_base = "https://xskt.com.vn/xsmb/ngay-"
year = [str(i) for i in range(2019,2021)]
month = [str(i) for i in range(1,13)]
date = [str(i) for i in range(1,31)]
for i in year:
  for j in month:
    for k in date:
      url = url_base + k + '-' + j + '-' + i
      print(url)

      page = urlopen(url)
      html_bytes = page.read()
      html = html_bytes.decode("utf-8")

      soup = BS(html)
      text = soup.get_text()

      word = text.split()
      if 'XSMB>' in word:
        index = word.index('XSMB>')
        print(int(word[index+5][5:7]))
        a = int(word[index+5][5:7])
        result.append(a)
      else:
        print("error")

print(result)        
```
